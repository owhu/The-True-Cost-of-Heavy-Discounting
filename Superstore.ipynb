```python
data['Year_Month'] = data['Order Date'].dt.to_period('M')

monthly_sales = data.groupby('Year_Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o', figsize=(12, 6), color='royalblue', linewidth=1)

plt.title('Monthly Sales Trend Over Time', fontsize=14, fontweight='bold', pad=10)
plt.xlabel('Date (Year-Month)', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(True, linestyle='-', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o', color='royalblue', linewidth=1)

plt.xticks(ticks=range(len(monthly_sales)), labels=monthly_sales.index.astype(str), rotation=90, fontsize=10)

plt.title('Monthly Sales Trend', fontsize=14, fontweight='bold', pad=10)
plt.xlabel('Date (Year-Month)', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
data['Is_Profitable'] = data['Profit'] >= 0

discount_analysis = data.groupby(['Sub-Category', 'Is_Profitable'])['Discount'].mean().unstack()
print(discount_analysis)
```

```python
heavy_discounts = data[data['Discount'] >= 0.5]

num_orders = len(heavy_discounts)
total_loss = heavy_discounts['Profit'].sum()

print(f"Number of transactions with 50%+ discounts: {num_orders}")
print(f"Total financial loss from these discounts: ${total_loss:,.2f}")
```

```python
losses_only = data[data['Profit'] < 0]

# Map Region vs. Customer Segment to see where the average discount is highest
discount_heat_map = losses_only.pivot_table(index='Region', 
                                            columns='Segment', 
                                            values='Discount', 
                                            aggfunc='mean')

print("--- Average Discount Rates on Unprofitable Orders ---")
print(discount_heat_map)
```

```python
central_losses = data[(data['Region'] == 'Central') & (data['Profit'] < 0)]

# Find the average discount and total loss by State
state_breakdown = central_losses.groupby('State').agg(
    Avg_Discount=('Discount', 'mean'),
    Total_Loss=('Profit', 'sum'),
    Order_Count=('Sales', 'count')
).sort_values(by='Total_Loss')

print(state_breakdown)
```

```python
shipping_drain = data.groupby('Ship Mode')[['Sales', 'Profit']].sum()
shipping_drain['Margin_%'] = (shipping_drain['Profit'] / shipping_drain['Sales']) * 100
print(shipping_drain)
```

```python
unprofitable_items = data[data['Profit'] < 0]

state_losses = unprofitable_items.groupby(['State', 'Sub-Category'])['Profit'].sum().sort_values()
print(state_losses.head(10))
```

```python
segment_analysis = data.groupby('Segment')[['Sales', 'Profit']].sum()
segment_analysis['Margin_%'] = (segment_analysis['Profit'] / segment_analysis['Sales']) * 100
print(segment_analysis)
```

```python
data.plot.scatter(x='Discount', y='Profit', alpha=0.5, figsize=(10, 6), color='crimson')
plt.title('How Discounts Kill Profitability')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import duckdb


# 1. Fix the 'period' data type issue by converting period columns to strings
for col in data.columns:
    if isinstance(data[col].dtype, pd.PeriodDtype):
        data[col] = data[col].astype(str)

# 2. Open a direct connection to DuckDB
conn = duckdb.connect(database=':memory:')

# 3. Hand your freshly cleaned Pandas DataFrame over to DuckDB
conn.register('data_table', data)

# 4. Write your exact query, referencing 'data_table'
query = """
WITH global_stats AS (
    SELECT 
        "Product Name", 
        AVG(Profit) AS avg_normal_profit
    FROM data_table
    WHERE Discount <= 0.10
    GROUP BY "Product Name"
)
SELECT 
    main."Product Name",
    main.Category,
    COUNT(main."Order ID") AS total_discounted_orders,
    ROUND(AVG(main.Discount), 2) AS avg_heavy_discount,
    ROUND(SUM(main.Profit), 2) AS total_financial_loss,
    ROUND(global_stats.avg_normal_profit, 2) AS benchmark_normal_profit
FROM data_table AS main
JOIN global_stats ON main."Product Name" = global_stats."Product Name"
WHERE main.Discount >= 0.50
  AND main.Profit < 0
GROUP BY main."Product Name", main.Category, global_stats.avg_normal_profit
ORDER BY total_financial_loss ASC
LIMIT 10;
"""

# 5. Execute the SQL and display
conn.execute(query).df()
```

```python
df_losses = conn.execute(query).df()

plt.figure(figsize=(12, 6))

sns.barplot(
    data=df_losses,
    x='total_financial_loss',
    y='Product Name',
    hue='Category',      # Color the bars based on their category
    palette='Reds_r'     # Uses a red color gradient (dark red for worst losses), _r means reversed so the worst offenders are darkest
)

plt.title('Top 10 Financial Bleeders: Total Loss Driven by Heavy Discounting (>=50%)', 
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Total Financial Loss ($)', fontsize=12)
plt.ylabel('Product Name', fontsize=12)

plt.tight_layout()

plt.show()
```
